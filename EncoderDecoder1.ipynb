```python
# ===============================================
# 📌 Headline Generation using LSTM Variants (Optimized)
# ===============================================

# STEP 1: Import Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Layer, Concatenate, GlobalAveragePooling1D, RepeatVector
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.meteor_score import meteor_score
try:
    from rouge_score import rouge_scorer
except ImportError:
    !pip install rouge-score
    from rouge_score import rouge_scorer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

# ===============================================
# STEP 2: Load Reduced Dataset (100 rows)
# ===============================================
try:
    # Load first 100 rows from each dataset
    df1 = pd.read_csv('news_summary.csv', encoding='ISO-8859-1', nrows=50)[['text', 'headlines']]
    df2 = pd.read_csv('news_summary_more.csv', encoding='ISO-8859-1', nrows=50)[['text', 'headlines']]
    df = pd.concat([df1, df2]).dropna().reset_index(drop=True)

    if df.empty:
        raise ValueError("The dataset is empty after loading.")
    print(f"Dataset loaded successfully with {len(df)} records.")
except Exception as e:
    print(f"Error loading dataset: {e}")
    # Create dummy data if files not found
    dummy_texts = ["This is sample news text about technology advancements."] * 50
    dummy_headlines = ["Tech advances rapidly"] * 50
    df = pd.DataFrame({'text': dummy_texts, 'headlines': dummy_headlines})
    print("Using dummy dataset for demonstration")

# ===============================================
# STEP 3: Clean Text
# ===============================================
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

df['clean_text'] = df['text'].apply(clean_text)
df['clean_headlines'] = df['headlines'].apply(clean_text)

# ===============================================
# STEP 4: Tokenization and Padding (Reduced Parameters)
# ===============================================
VOCAB_SIZE = 2000  # Reduced for smaller dataset
MAX_INPUT_LEN = 30  # Reduced max length
MAX_TARGET_LEN = 8  # Reduced max length

input_tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token="<OOV>")
input_tokenizer.fit_on_texts(df['clean_text'])
input_seq = input_tokenizer.texts_to_sequences(df['clean_text'])
input_seq = pad_sequences(input_seq, maxlen=MAX_INPUT_LEN, padding='post')

target_tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token="<OOV>")
target_tokenizer.fit_on_texts(df['clean_headlines'])
target_seq = target_tokenizer.texts_to_sequences(df['clean_headlines'])
target_seq = pad_sequences(target_seq, maxlen=MAX_TARGET_LEN, padding='post')

decoder_input_seq = target_seq[:, :-1]
decoder_target_seq = target_seq[:, 1:]
decoder_target_seq = np.expand_dims(decoder_target_seq, -1)

# Split data (80% train, 20% validation)
X_train, X_val, Xd_train, Xd_val, Y_train, Y_val = train_test_split(
    input_seq, decoder_input_seq, decoder_target_seq, test_size=0.2, random_state=42
)

# ===============================================
# STEP 5A: Vanilla LSTM Encoder-Decoder (Simplified)
# ===============================================
def build_lstm_model():
    # Reduced embedding and LSTM units for smaller dataset
    encoder_inputs = Input(shape=(MAX_INPUT_LEN,))
    enc_emb = Embedding(VOCAB_SIZE, 64)(encoder_inputs)
    enc_out, state_h, state_c = LSTM(64, return_state=True)(enc_emb)

    decoder_inputs = Input(shape=(MAX_TARGET_LEN - 1,))
    dec_emb = Embedding(VOCAB_SIZE, 64)(decoder_inputs)
    dec_lstm = LSTM(64, return_sequences=True)(dec_emb, initial_state=[state_h, state_c])

    output = Dense(VOCAB_SIZE, activation='softmax')(dec_lstm)
    model = Model([encoder_inputs, decoder_inputs], output)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# ===============================================
# STEP 5B: LSTM with Bahdanau Attention (Simplified)
# ===============================================
class BahdanauAttention(Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = Dense(units)
        self.W2 = Dense(units)
        self.V = Dense(1)

    def call(self, query, values):
        query = tf.expand_dims(query, 1)
        score = self.V(tf.nn.tanh(self.W1(query) + self.W2(values)))
        attention_weights = tf.nn.softmax(score, axis=1)
        context_vector = attention_weights * values
        context_vector = tf.reduce_sum(context_vector, axis=1)
        return context_vector, attention_weights

def build_attention_model():
    # Reduced units for smaller dataset
    encoder_inputs = Input(shape=(MAX_INPUT_LEN,))
    enc_emb = Embedding(VOCAB_SIZE, 64)(encoder_inputs)
    enc_out, state_h, state_c = LSTM(64, return_sequences=True, return_state=True)(enc_emb)

    decoder_inputs = Input(shape=(MAX_TARGET_LEN - 1,))
    dec_emb = Embedding(VOCAB_SIZE, 64)(decoder_inputs)
    dec_lstm_out = LSTM(64, return_sequences=True)(dec_emb, initial_state=[state_h, state_c])

    attention = BahdanauAttention(64)
    context_vector, _ = attention(state_h, enc_out)
    repeated_context = RepeatVector(MAX_TARGET_LEN - 1)(context_vector)
    concat = Concatenate(axis=-1)([dec_lstm_out, repeated_context])
    output = Dense(VOCAB_SIZE, activation='softmax')(concat)

    model = Model([encoder_inputs, decoder_inputs], output)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# ===============================================
# STEP 5C: LSTM with Self-Attention (Simplified)
# ===============================================
class SelfAttention(Layer):
    def __init__(self, units):
        super(SelfAttention, self).__init__()
        self.Wq = Dense(units)
        self.Wk = Dense(units)
        self.Wv = Dense(units)

    def call(self, x):
        q = self.Wq(x)
        k = self.Wk(x)
        v = self.Wv(x)

        score = tf.matmul(q, k, transpose_b=True) / tf.math.sqrt(tf.cast(tf.shape(k)[-1], tf.float32))
        weights = tf.nn.softmax(score, axis=-1)
        output = tf.matmul(weights, v)
        return output

def build_self_attention_model():
    # Reduced units for smaller dataset
    encoder_inputs = Input(shape=(MAX_INPUT_LEN,))
    enc_emb = Embedding(VOCAB_SIZE, 64)(encoder_inputs)
    enc_out, state_h, state_c = LSTM(64, return_sequences=True, return_state=True)(enc_emb)

    encoder_sa = SelfAttention(64)(enc_out)
    encoder_pooled = GlobalAveragePooling1D()(encoder_sa)
    encoder_repeated = RepeatVector(MAX_TARGET_LEN - 1)(encoder_pooled)

    decoder_inputs = Input(shape=(MAX_TARGET_LEN - 1,))
    dec_emb = Embedding(VOCAB_SIZE, 64)(decoder_inputs)
    dec_lstm_out = LSTM(64, return_sequences=True)(dec_emb, initial_state=[state_h, state_c])

    concat = Concatenate(axis=-1)([dec_lstm_out, encoder_repeated])
    output = Dense(VOCAB_SIZE, activation='softmax')(concat)

    model = Model([encoder_inputs, decoder_inputs], output)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# ===============================================
# STEP 6: Train All Models (Reduced Epochs)
# ===============================================
print("\nTraining Vanilla LSTM...")
vanilla_model = build_lstm_model()
vanilla_model.summary()
history_vanilla = vanilla_model.fit(
    [X_train, Xd_train], Y_train,
    validation_data=([X_val, Xd_val], Y_val),
    batch_size=32, epochs=5, verbose=1
)

print("\nTraining LSTM with Attention...")
attention_model = build_attention_model()
attention_model.summary()
history_attention = attention_model.fit(
    [X_train, Xd_train], Y_train,
    validation_data=([X_val, Xd_val], Y_val),
    batch_size=32, epochs=5, verbose=1
)

print("\nTraining LSTM with Self-Attention...")
self_attention_model = build_self_attention_model()
self_attention_model.summary()
history_self_attention = self_attention_model.fit(
    [X_train, Xd_train], Y_train,
    validation_data=([X_val, Xd_val], Y_val),
    batch_size=32, epochs=5, verbose=1
)

# ===============================================
# STEP 7: Comparison Plot
# ===============================================
def plot_model_comparison(histories, labels):
    plt.figure(figsize=(12, 6))

    # Training Loss
    for history, label in zip(histories, labels):
        plt.plot(history.history['loss'], label=f'{label} - Train')

    # Validation Loss
    for history, label in zip(histories, labels):
        plt.plot(history.history['val_loss'], linestyle='--', label=f'{label} - Val')

    plt.title('Model Comparison (100 samples)')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_model_comparison(
    [history_vanilla, history_attention, history_self_attention],
    ['Vanilla LSTM', 'LSTM + Attention', 'LSTM + Self-Attention']
)

# ===============================================
# STEP 8: Inference Function
# ===============================================
def generate_headline(input_text, model, input_tokenizer, target_tokenizer):
    input_seq = input_tokenizer.texts_to_sequences([clean_text(input_text)])
    input_seq = pad_sequences(input_seq, maxlen=MAX_INPUT_LEN, padding='post')

    decoder_input = np.zeros((1, MAX_TARGET_LEN - 1))
    decoded_words = []

    for i in range(MAX_TARGET_LEN - 1):
        prediction = model.predict([input_seq, decoder_input], verbose=0)
        sampled_token_index = np.argmax(prediction[0, i, :])
        sampled_word = target_tokenizer.index_word.get(sampled_token_index, '')
        if sampled_word == '' or sampled_word == '<OOV>':
            break
        decoded_words.append(sampled_word)
        if i < MAX_TARGET_LEN - 2:
            decoder_input[0, i] = sampled_token_index

    return ' '.join(decoded_words)

# ===============================================
# STEP 9: Example Usage
# ===============================================
test_text = "NASA has launched a new satellite to study climate change"
print("\nGenerated Headlines:")
print("Vanilla LSTM:", generate_headline(test_text, vanilla_model, input_tokenizer, target_tokenizer))
print("Attention LSTM:", generate_headline(test_text, attention_model, input_tokenizer, target_tokenizer))
print("Self-Attention LSTM:", generate_headline(test_text, self_attention_model, input_tokenizer, target_tokenizer))

# ===============================================
# STEP 10: Enhanced Evaluation Metrics with Visualizations
# ===============================================
def evaluate_models(models, model_names, X_val, Xd_val, Y_val, input_tokenizer, target_tokenizer):
    results = {
        'BLEU-1': [], 'BLEU-4': [],
        'ROUGE-L': [], 'METEOR': [],
        'Accuracy': []
    }

    # Generate predictions for each model
    all_predictions = []
    for model in models:
        predictions = []
        for i in range(len(X_val)):
            input_seq = X_val[i:i+1]
            decoder_input = np.zeros((1, MAX_TARGET_LEN - 1))

            decoded_words = []
            for j in range(MAX_TARGET_LEN - 1):
                prediction = model.predict([input_seq, decoder_input], verbose=0)
                sampled_token_index = np.argmax(prediction[0, j, :])
                sampled_word = target_tokenizer.index_word.get(sampled_token_index, '')
                if sampled_word == '' or sampled_word == '<OOV>':
                    break
                decoded_words.append(sampled_word)
                if j < MAX_TARGET_LEN - 2:
                    decoder_input[0, j] = sampled_token_index

            predictions.append(' '.join(decoded_words))
        all_predictions.append(predictions)

    # Get reference headlines
    reference_strings = []
    reference_tokens = []
    for seq in Y_val:
        ref_words = []
        for token in seq:
            if token[0] > 0:  # Skip padding
                word = target_tokenizer.index_word.get(token[0], '')
                if word and word != '<OOV>':
                    ref_words.append(word)
        reference_strings.append(' '.join(ref_words))
        reference_tokens.append(ref_words)

    # Calculate metrics for each model
    for i, (model, preds) in enumerate(zip(models, all_predictions)):
        # BLEU Scores
        bleu1_scores = []
        bleu4_scores = []
        for ref_tokens, pred in zip(reference_tokens, preds):
            pred_tokens = pred.split()
            if len(ref_tokens) > 0 and len(pred_tokens) > 0:  # Ensure non-empty sequences
                bleu1_scores.append(sentence_bleu([ref_tokens], pred_tokens, weights=(1, 0, 0, 0)))
                bleu4_scores.append(sentence_bleu([ref_tokens], pred_tokens, weights=(0.25, 0.25, 0.25, 0.25)))

        results['BLEU-1'].append(np.mean(bleu1_scores) if bleu1_scores else 0)
        results['BLEU-4'].append(np.mean(bleu4_scores) if bleu4_scores else 0)

        # ROUGE Score
        try:
            scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
            rouge_scores = [scorer.score(ref, pred)['rougeL'] for ref, pred in zip(reference_strings, preds)]
            results['ROUGE-L'].append(np.mean([score.fmeasure for score in rouge_scores]))
        except Exception as e:
            print(f"ROUGE score calculation failed: {e}, using 0 as default")
            results['ROUGE-L'].append(0)

        # METEOR Score
        meteor_scores = []
        for ref_tokens, pred in zip(reference_tokens, preds):
            pred_tokens = pred.split()
            if len(ref_tokens) > 0 and len(pred_tokens) > 0:  # Ensure non-empty sequences
                try:
                    meteor_scores.append(meteor_score([ref_tokens], pred_tokens))
                except:
                    meteor_scores.append(0)
        results['METEOR'].append(np.mean(meteor_scores) if meteor_scores else 0)

        # Accuracy (Exact Match)
        exact_matches = [1 if pred == ref else 0 for pred, ref in zip(preds, reference_strings)]
        results['Accuracy'].append(np.mean(exact_matches))

    # Create dataframe for results
    metrics_df = pd.DataFrame(results, index=model_names)

    # Plotting
    plt.figure(figsize=(15, 10))

    # BLEU Scores Plot
    plt.subplot(2, 2, 1)
    metrics_df[['BLEU-1', 'BLEU-4']].plot(kind='bar', ax=plt.gca())
    plt.title('BLEU Scores Comparison')
    plt.ylabel('Score')
    plt.xticks(rotation=0)

    # ROUGE-L Plot
    plt.subplot(2, 2, 2)
    metrics_df['ROUGE-L'].plot(kind='bar', color='green', ax=plt.gca())
    plt.title('ROUGE-L F1 Score Comparison')
    plt.ylabel('Score')
    plt.xticks(rotation=0)

    # METEOR Plot
    plt.subplot(2, 2, 3)
    metrics_df['METEOR'].plot(kind='bar', color='orange', ax=plt.gca())
    plt.title('METEOR Score Comparison')
    plt.ylabel('Score')
    plt.xticks(rotation=0)

    # Accuracy Plot
    plt.subplot(2, 2, 4)
    metrics_df['Accuracy'].plot(kind='bar', color='red', ax=plt.gca())
    plt.title('Exact Match Accuracy Comparison')
    plt.ylabel('Accuracy')
    plt.xticks(rotation=0)

    plt.tight_layout()
    plt.show()

    return metrics_df

# ===============================================
# STEP 11: Run Evaluation and Show Results
# ===============================================
models = [vanilla_model, attention_model, self_attention_model]
model_names = ['Vanilla LSTM', 'LSTM + Attention', 'LSTM + Self-Attention']

metrics_results = evaluate_models(
    models, model_names, X_val, Xd_val, Y_val,
    input_tokenizer, target_tokenizer
)

print("\nDetailed Metrics Results:")
print(metrics_results)

# ===============================================
# STEP 12: Sample Predictions Comparison
# ===============================================
def show_sample_predictions(models, model_names, samples=3):
    for i in range(samples):
        idx = np.random.randint(0, len(X_val))
        input_text = ' '.join([input_tokenizer.index_word.get(token, '') for token in X_val[idx] if token > 0])
        actual_headline = ' '.join([target_tokenizer.index_word.get(token[0], '') for token in Y_val[idx] if token[0] > 0])

        print(f"\nSample {i+1}:")
        print(f"Input Text: {input_text}")
        print(f"Actual Headline: {actual_headline}")

        for model, name in zip(models, model_names):
            pred = generate_headline(input_text, model, input_tokenizer, target_tokenizer)
            print(f"{name}: {pred}")

show_sample_predictions(models, model_names)
```
